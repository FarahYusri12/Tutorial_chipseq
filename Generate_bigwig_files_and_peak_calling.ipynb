```python
# generate bigwig files (loop)
%%bash
for bam in *.sorted.bam; do
out=${bam/.sorted.bam}.bw
bamCoverage \
--bam "$bam" \
--normalizeUsing RPKM \
--extendReads 200 \
-o "$out"
done
```

```python
#mount bam file to contenr
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810900.sorted.bam" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810900.sorted.bam.bai" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810907.sorted.bam" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810907.sorted.bam.bai" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810912.sorted.bam" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810912.sorted.bam.bai" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810918.sorted.bam" .
!cp "/content/drive/MyDrive/portfolio/Chip-seq tutorial/SRR1810918.sorted.bam.bai" .
```

```python
#SRR1810900(treatment), SRR1810912(control)
%%bash
macs3 callpeak \
-t SRR1810900.sorted.bam \
-c SRR1810912.sorted.bam \
-f BAM \
-g hs \
-n SRR1810900 \
--outdir SRR1810900_peak \
--keep-dup all \
--call-summits
```

```python
#SRR1810907(treatment), SRR1810912(control)
%%bash
macs3 callpeak \
-t SRR1810907.sorted.bam \
-c SRR1810912.sorted.bam \
-f BAM \
-g hs \
-n SRR1810907 \
--outdir SRR1810907_peak \
--keep-dup all \
--call-summits
```

```python
#SRR1810918(treatment), SRR1810912(control)
%%bash
macs3 callpeak \
-t SRR1810918.sorted.bam \
-c SRR1810912.sorted.bam \
-f BAM \
-g hs \
-n SRR1810918 \
--outdir SRR1810918_peak \
--keep-dup all \
--call-summits
```
